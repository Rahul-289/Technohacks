```python
import pandas as pd
import numpy as np

# Load the Iris dataset from CSV
file_path = r'C:\Users\LENOVO\Downloads\Iris (1).csv'
iris_data = pd.read_csv(file_path)

# Display the first few rows of the dataset
print("First few rows of the dataset:")
print(iris_data.head())

# Calculate summary statistics for each numerical column
summary_stats = iris_data.describe()

# Display the calculated summary statistics
print("\nSummary Statistics:")
print(summary_stats)

# Calculate mode for each column
modes = iris_data.mode(axis=0)
print("\nMode:")
print(modes)

# Calculate mean for each column
means = iris_data.mean(axis=0)
print("\nMean:")
print(means)

# Calculating standard deviation using NumPy for each column
std_dev_np = np.std(iris_data, axis=0)
print("\nStandard Deviation (using NumPy):")
print(std_dev_np)
```
